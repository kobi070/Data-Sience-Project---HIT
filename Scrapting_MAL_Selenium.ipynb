```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import numpy as np
import re
from collections import Counter
```

```python
PATH = "C:/Users/Kobi/Downloads/chromedriver.exe"
driver = webdriver.Chrome(PATH)
wait = WebDriverWait(driver, 10)
```

```python
### get all top anime links ###
def get_all_top_links(site_url):
    all_top_urls = []
    for i in range(0,10000,50):
        all_top_urls.append(site_url+'%3Flimit%3D50&limit='+str(i))
    return all_top_urls
```

```python
### Site's url's ###
mal_genre_front_page = 'https://myanimelist.net/anime.php'
top_anime_url = 'https://myanimelist.net/topanime.php?type=tv'
# ### All genre Links list ###
# mal_genre_links = get_all_genre_links(mal_genre_front_page,genre_dict)
### All Top 0-19,000 Anime's in mal URL list ###
top_anime_list = get_all_top_links(top_anime_url)
display(top_anime_list)
df = pd.read_csv('top_anime_df_v3.csv')

```

```python
links = []
for page in top_anime_list:
    print(page)
    driver.get(page)
    link = driver.find_elements(By.CSS_SELECTOR,'div[class="detail"] h3 a ')
    for item in link:
        links.append(item.get_attribute('href'))
```

```python
ratings = []
generes = []
sources = []
studios = []
premiereds = []
types = []


for i_link in links:
    
    driver.get(i_link)
    try:
        pre = wait.until(EC.visibility_of_element_located((By.XPATH,'//*[contains(text(),"Premiered:")]/parent::div')))
        premiereds.append(pre.text)
    except:
        print(f" An Error As Occoured at  - Premier !")
        premiereds.append(np.nan)
        continue
    
    try:
        source = wait.until(EC.visibility_of_element_located((By.XPATH,'//*[contains(text(),"Source:")]/parent::div')))
        sources.append(source.text)
    except:
        print(f" An Error As Occoured at  - Source !")
        sources.append(np.nan)
        continue
    try:
        studio = wait.until(EC.visibility_of_element_located((By.XPATH,'//*[contains(text(),"Studios:")]/parent::div')))
        studios.append(studio.text)
    except:
        print(f" An Error As Occoured at  - Studios !")
        studios.append(np.nan)
        continue

    try:
        genere = wait.until(EC.visibility_of_element_located((By.XPATH,'//*[contains(text(),"Genre")]/parent::div')))
        generes.append(genere.text)
    except:
        print(f" An Error As Occoured at  - Generes !")
        generes.append(np.nan)
        continue
        
    try:
        rating = wait.until(EC.visibility_of_element_located((By.XPATH,'//*[contains(text(),"Rating:")]/parent::div')))
        ratings.append(rating.text)
    except:
        print(f" An Error As Occoured at  - Rating !")
        mean = np.mean(ratings)
        ratings.append(mean)
    
    time.sleep(5)

driver.quit()
        
        
        
                                                         
```

```python
def fix_string_unkown(listt,mode):
    
    ### Fix String Values of Source and Studio ###
    result = []
    for i,_ in enumerate(listt,start=1):
        if len(listt) != len(result):
            if listt[i] != 'unkown':
                try:
                    a = str(listt[i])
                    b = a.split(':')[1]
                    result.append(b)
                    
                except:pass
            else:
                a = mode
                result.append(a)
        else: break
    ### If Len of lists is not a match keep filling by listt[i] ####
    if len(result) != len(listt):
        for i in range(len(result),len(listt),1):
            a = mode
            result.append(a)

    return result

def fix_string_nan(listt,mode):
    
    ### Fix Value By NaN ###
    result = []
    for i,_ in enumerate(listt,start=1):
        if listt[i] != np.nan:
            try:
                a = str(listt[i])
                b = a.split(':')[1]
                result.append(b)
            except: 
                a = mode
                result.append(a)
        else:
            a = mode
            result.append(a)
    return result

def list_wo_unkown_nan(listt):
    result = []
    for i,_ in enumerate(listt,start=1):
        if listt[i] != 'unkown' or listt[i] != np.nan:
            try:
                a = str(listt[i])
                b = a.split(':')[1]
                result.append(b)
                print(b)
            except:
                continue
        else:continue
    return result
def fix_list_no_nan(dataframe,col_name,mode):
    result = []
    listt = dataframe[col_name].to_list()
    for i in listt:
        if i is np.nan:
            result.append(mode)
        else:
            result.append(i)
    return result
        
def list_wo_unkown_nan_nostr(dataframe, col_name):
    
    dataframe.dropna(how='any',inplace=True)
    result = dataframe[col_name].to_list()
    f3 = Counter(result)
    mode_ep = f3.most_common(1)
    
    
    return result, mode_ep[0][0]
```

```python
animeCSV_f.drop(axis=1,columns='Unnamed: 0')
animeCSV_f_s = pd.read_csv('Not_Done.csv',index_col='Rank')
animeCSV_f_s.drop(axis=1,columns='Unnamed: 0')



    
    

fixed_source_mode = list_wo_unkown_nan(animeCSV_f_s['Source'])
fixed_studio_mode =list_wo_unkown_nan(animeCSV_f['Studio'])


f1 = Counter(fixed_source_mode)
f2 = Counter(fixed_studio_mode)
mode_source = f1.most_common(1)
mode_studio = f2.most_common(1)
fixed_source = fix_string_unkown(animeCSV_f_s['Source'],mode_source[0][0])
fixed_studio = fix_string_nan(animeCSV_f['Studio'],mode_studio[0][0])


animeCSV_f_s2 = animeCSV_f_s.copy()


animeCSV_f_s2['Source'] = fixed_source
animeCSV_f_s2['Studio'] = fixed_studio



animeCSV_f_s2['Premiered'] = animeCSV_f_s2['Start Date']
animeCSV_f_s2.drop(axis=1,columns=['Start Date','Finish Date','URL','Generes','Unnamed: 0'],inplace=True)



    

animeCSV_f_s2


```
